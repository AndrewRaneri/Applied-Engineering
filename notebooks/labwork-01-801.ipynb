### <span style="color:blue">Create a Python code to compute the area of a right triangle with sides $a = 1.23$ and $b = 2.45$, and its hypotenuse. On a Markdown cell, explain your results mathematically.</span>

````python
import math
a = 1.23
b = 2.45
area = a * b / 2.0
print('area = ',area)
h = math.sqrt(a**2+b**2)
print('hypotenuse = ', h)
````
